```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("https://files.grouplens.org/datasets/movielens/ml-100k/u.data")
df
```

```python
df = pd.read_csv("https://files.grouplens.org/datasets/movielens/ml-100k/u.data", names=["user_id", "item_id", "review", "timestamp"], sep="\t")
df
```

```python
df = pd.pivot_table(df.drop("timestamp", axis=1), index="user_id", columns="item_id", values="review")
df
```

```python
df = df.fillna(0)
df
```

```python
## cos類似度をnumpyを使って計算


dot = np.dot(df.iloc[0], df.iloc[1])

norm_a = np.linalg.norm(df.iloc[0])
norm_b = np.linalg.norm(df.iloc[1])

dot / (norm_a*norm_b)
```

```python
cosine_similarity([df.iloc[0]], [df.iloc[1]])
```

```python
cos_list = []
for user_id in df.index:
  cos_list.append(cosine_similarity([df.loc[2, 2:]], [df.loc[user_id, 2:]]))
cos_list
```
